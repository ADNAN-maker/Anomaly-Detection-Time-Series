Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
***
Ans:- Anomaly Detection is the process of identifying data points, events, or observations that deviate significantly from the normal or expected behavior. These deviations are often referred to as anomalies, outliers, or novelties.

There are three main types of anomalies:

*   **Point Anomalies:** A single data instance is anomalous if it's far off from the rest of the data.
    *   *Example:* A sudden, unusually high transaction amount in a customer's credit card history.
*   **Contextual Anomalies:** A data instance is considered anomalous within a specific context, but not necessarily otherwise.
    *   *Example:* Spending $100 on groceries is normal, but spending $1000 on groceries in a single day might be a contextual anomaly.
*   **Collective Anomalies:** A collection of related data instances is anomalous, even if individual instances within the collection are not anomalous by themselves.
    *   *Example:* A sudden drop in website traffic followed by a gradual increase might not be anomalous individually, but the combination could indicate a server issue or a denial-of-service attack.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
***
Ans:- 1. Isolation Forest (iForest)

* Approach:

  * Ensemble-based algorithm.

  * Builds multiple decision trees by randomly selecting features and split values.

  * Anomalies are easier to ‚Äúisolate‚Äù (fewer splits needed), while normal points require more splits.

  * Outputs an anomaly score per point.

* Suitable Use Cases:

  * Works well with high-dimensional datasets.

  * Handles large datasets efficiently (linear complexity).

  * Best for global outlier detection (points far from the bulk of the data).

  * Examples: fraud detection, network intrusion detection, manufacturing defect detection.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

* Approach:

  * Density-based clustering algorithm.

  * Groups data into clusters of high density separated by low-density regions.

  * Points that do not belong to any dense cluster are treated as outliers/noise.

  * Requires two parameters: Œµ (neighborhood size) and minPts (minimum points in neighborhood).

* Suitable Use Cases:

  * Good when data has clusters of varying shapes and sizes.

  * Identifies local outliers as points not fitting into clusters.

  * Not ideal for high-dimensional data (curse of dimensionality affects distance metrics).

  * Examples: spatial data analysis, geographic clustering, image segmentation with outlier detection.

3. Local Outlier Factor (LOF)

* Approach:

  * Density-based anomaly detection.

  * Compares the local density of a point to the densities of its neighbors.

  * A point is an outlier if it has a significantly lower density than its neighbors.

  * Outputs a relative anomaly score (higher = more anomalous).

* Suitable Use Cases:

  * Detects local outliers (that may be normal globally but unusual within a neighborhood).

  * Sensitive to parameter choice (k = number of neighbors).

  * Works well with datasets with varying density regions.

  * Examples: credit card fraud (locally unusual transactions), sensor network anomalies, rare event detection.

Question 3: What are the key components of a Time Series? Explain each with one example.
***
Ans:- A time series is a sequence of data points indexed in time order. The key components of a time series are:

*   **Trend:** The long-term movement of the series. It represents the underlying direction of the data over time, ignoring short-term fluctuations.
    *   *Example:* The increasing trend in the global average temperature over the past century.
*   **Seasonality:** The repeating patterns or cycles that occur at regular intervals, such as daily, weekly, monthly, or yearly.
    *   *Example:* The increase in ice cream sales during summer months each year.
*   **Cyclicity:** Patterns that repeat over longer, irregular periods, often influenced by economic conditions or other factors. Unlike seasonality, cycles don't have a fixed frequency.
    *   *Example:* The boom and bust cycles in the stock market.
*   **Irregularity (or Noise):** Random fluctuations or unpredictable events that are not explained by trend, seasonality, or cyclicity.
    *   *Example:* A sudden spike in electricity consumption due to an unexpected heatwave.

Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?
***
Ans:- 1. Definition of Stationarity

A stationary time series is one whose statistical properties (mean, variance, autocorrelation, etc.) are constant over time.
* In other words, the series does not exhibit trends, seasonality, or changing variance.
* Stationarity is important because many forecasting models (like ARIMA) assume the input data is stationary.

Types of Stationarity:

I. Strict stationarity: Full distribution of the series is constant over time.

II. Weak (or covariance) stationarity: Mean, variance, and autocovariance are constant over time (most practical definition).

2. How to Test for Stationarity

Common methods include:
* Visual Inspection: Plot the series ‚Äî look for trends, seasonality, or changing variance.
* Summary Statistics: Split the series into parts and compare mean/variance.
* Statistical Tests:

  I. ADF (Augmented Dickey-Fuller) Test:
  * Null Hypothesis (H‚ÇÄ): Series is non-stationary.
  * If p-value < 0.05 ‚Üí reject H‚ÇÄ ‚Üí series is stationary.

  II. KPSS (Kwiatkowski-Phillips-Schmidt-Shin) Test:
  * Null Hypothesis (H‚ÇÄ): Series is stationary.
  * If p-value < 0.05 ‚Üí reject H‚ÇÄ ‚Üí series is non-stationary.

  III. Often both ADF and KPSS are used together for confirmation.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.
***

Ans:- Here's a differentiation between AR, MA, ARIMA, SARIMA, and SARIMAX models:

| Model   | Structure                                                                                                | Application                                                                                                |
|---------|----------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------|
| **AR(p)** | **Autoregressive Model:** Predicts future values based on a linear combination of past values of the series. The 'p' denotes the order (number of past values considered). | Suitable for time series where the current value is dependent on past values. Example: Stock prices, temperature readings. |
| **MA(q)** | **Moving Average Model:** Predicts future values based on a linear combination of past forecast errors (residuals). The 'q' denotes the order (number of past forecast errors considered). | Suitable for time series where the current value is dependent on past random shocks or errors. Example: White noise series, data with short-term dependencies. |
| **ARIMA(p, d, q)** | **Autoregressive Integrated Moving Average:** Combines AR and MA models with differencing. 'p' is the AR order, 'd' is the degree of differencing (to make the series stationary), and 'q' is the MA order. | Suitable for non-stationary time series that can be made stationary by differencing. Widely used for forecasting in various domains like finance, economics, and weather. |
| **SARIMA(p, d, q)(P, D, Q)m** | **Seasonal Autoregressive Integrated Moving Average:** Extends ARIMA to handle seasonality. (p, d, q) are non-seasonal orders, (P, D, Q) are seasonal orders, and 'm' is the number of periods in each season. | Suitable for time series with clear seasonal patterns. Example: Retail sales (monthly or quarterly), energy consumption (daily or hourly). |
| **SARIMAX(p, d, q)(P, D, Q)m, X** | **Seasonal Autoregressive Integrated Moving Average with eXogenous variables:** Extends SARIMA by including external or exogenous variables that can influence the time series. | Suitable for time series with seasonality and where external factors play a significant role in the series' behavior. Example: Predicting sales influenced by advertising spend, forecasting temperature influenced by humidity. |

**Key Differences:**

*   **Stationarity:** AR and MA models assume stationarity, while ARIMA and its extensions (SARIMA, SARIMAX) can handle non-stationary data through differencing.
*   **Seasonality:** SARIMA and SARIMAX explicitly model seasonality, while AR, MA, and ARIMA do not.
*   **Exogenous Variables:** Only SARIMAX incorporates external variables into the model.

Choosing the appropriate model depends on the characteristics of the time series data, including the presence of trends, seasonality, and the influence of external factors. Analyzing the autocorrelation function (ACF) and partial autocorrelation function (PACF) of the time series can help in determining the appropriate orders (p, q, P, Q) for these models.

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

Dataset:
* NYC Taxi Fare Data
* AirPassengers Dataset
***

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm

# Load the AirPassengers dataset
data = sm.datasets.get_rdataset("AirPassengers").data
data['Month'] = pd.to_datetime(data['time'], format='%Y')
data.set_index('Month', inplace=True)

# Rename the column for clarity
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Plot original series
plt.figure(figsize=(10, 5))
plt.plot(data['Passengers'], label='AirPassengers')
plt.title("AirPassengers Dataset - Monthly Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
# Perform decomposition (multiplicative, since variance increases with time)
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

# Plot decomposition
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
***

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np

# Load the NYC Taxi Fare dataset
# You might need to download this dataset or use a path to where it's stored
# For demonstration, let's create a synthetic dataset resembling taxi fare data
np.random.seed(42)
n_samples = 300
X = np.random.rand(n_samples, 2) * 10 # Feature 1 (e.g., distance) and Feature 2 (e.g., time)

# Introduce some anomalies
X[0, :] = [20, 25] # Anomaly 1 (high distance, high time)
X[1, :] = [0.5, 15] # Anomaly 2 (low distance, high time)
X[2, :] = [18, 1] # Anomaly 3 (high distance, low time)

# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])

# Apply Isolation Forest
model = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
model.fit(df)

# Predict anomalies (-1 for outliers, 1 for inliers)
df['anomaly'] = model.predict(df)

# Visualize the anomalies
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Feature_1'], df['Feature_2'], c=df['anomaly'], cmap='coolwarm', s=20)
plt.title('Anomaly Detection using Isolation Forest (Synthetic Data)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(scatter, label='Anomaly (-1: Outlier, 1: Inlier)')
plt.grid(True)
plt.show()

# Print the anomalies
print("Detected Anomalies:")
print(df[df['anomaly'] == -1])
```

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results
***

```python
# 1. Import libraries and load dataset
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = sm.datasets.get_rdataset("AirPassengers").data
data['Month'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Plot the data
plt.figure(figsize=(10,5))
plt.plot(data['Passengers'], label="AirPassengers")
plt.title("Monthly Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()

# 2. Fit a SARIMA model
#AirPassengers is monthly data with yearly seasonality (12 months).
#A common order used in literature: SARIMA(1,1,1)(1,1,1,12)

# Define SARIMA model
model = SARIMAX(data['Passengers'],
                order=(1,1,1),        # (p,d,q)
                seasonal_order=(1,1,1,12),  # (P,D,Q,s)
                enforce_stationarity=False,
                enforce_invertibility=False)

# Fit the model
results = model.fit(disp=False)

print(results.summary())

# 3. Forecast the next 12 months

forecast = results.get_forecast(steps=12)

# Get confidence intervals
forecast_ci = forecast.conf_int()

# Plot results
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Passengers'], label='Observed')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)

plt.title("SARIMA Forecast - AirPassengers")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
***

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

# Create a synthetic dataset resembling numerical data for anomaly detection
np.random.seed(42)
X = np.random.rand(300, 2) * 10  # Generate 300 samples with 2 features

# Introduce some anomalies
X[0, :] = [20, 25]  # Anomaly 1
X[1, :] = [0.5, 15] # Anomaly 2
X[2, :] = [18, 1]   # Anomaly 3
X[3, :] = [10, 18]  # Anomaly 4

# Apply Local Outlier Factor (LOF)
# n_neighbors: Number of neighbors to consider
# contamination: The proportion of outliers in the data set. 'auto' is a good starting point.
lof = LocalOutlierFactor(n_neighbors=20, contamination='auto')
y_pred = lof.fit_predict(X)

# Anomalies are indicated by -1, inliers by 1
anomalies = X[y_pred == -1]
inliers = X[y_pred == 1]

# Visualize the anomalies
plt.figure(figsize=(10, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], label='Inliers', c='blue', s=20)
plt.scatter(anomalies[:, 0], anomalies[:, 1], label='Anomalies', c='red', s=20)
plt.title('Anomaly Detection using Local Outlier Factor (LOF) - Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# Print the detected anomalies
print("Detected Anomalies:")
print(anomalies)
```

Here's a potential real-time data science workflow for forecasting energy demand and detecting anomalies in streaming data for a power grid monitoring company:

    *   **Isolation Forest:** It's generally efficient and scales well to high-dimensional data. It can quickly identify anomalies without needing to define a density threshold explicitly. It would be applied to a sliding window of recent data. Anomalies would be flagged based on their isolation score.
    *   **Local Outlier Factor (LOF):** LOF is good for detecting local anomalies, which might be relevant in a power grid where anomalies could be localized to specific regions or conditions. It requires defining a neighborhood size.

3.  **Time Series Modeling for Short-term Forecasting:**
    *   Given the presence of seasonality (daily, weekly, yearly patterns in energy usage) and the influence of external factors (weather conditions), **SARIMAX** would be the most appropriate time series model for short-term forecasting.

        *   **Forecast Accuracy:** Continuously monitor the forecasting accuracy of the SARIMAX model in real-time against actual energy usage.
